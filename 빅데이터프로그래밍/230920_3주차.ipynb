```python
import pandas as pd
```

```python
df = pd.read_csv('/content/owid-covid-data.csv')
```

```python
cols = ['iso_code', 'location', 'date', 'new_cases', 'population']
df = df[cols]

df.head()
```

```python
kor = df[df['location'] == 'South Korea']
kor = kor.rename(columns={'new_cases' : 'KOR'})
kor.head()
```

```python
kor.head(300)
```

```python
usa = df[df['location']=='United States']
usa = usa.rename(columns={'new_cases' : 'USA'})
usa.head()
```

```python
fra = df[df['location']=='France']
fra = fra.rename(columns={'new_cases' : 'FRA'})
fra.head()
```

```python
icl = df[df['location']=='Iceland']
icl = icl.rename(columns={'new_cases' : 'ICL'})
icl.head()
```

```python
nrw = df[df['location']=='Norway']
nrw = nrw.rename(columns={'new_cases' : 'NRW'})
nrw.head()
```

```python
kor_pop = kor[kor['date']=='2020-01-22']['population'].iloc[0]
usa_pop = usa[usa['date']=='2020-01-22']['population'].iloc[0]
fra_pop = fra[fra['date']=='2020-01-22']['population'].iloc[0]
icl_pop = icl[icl['date']=='2020-01-22']['population'].iloc[0]
nrw_pop = nrw[nrw['date']=='2020-01-22']['population'].iloc[0]

print(kor_pop,usa_pop,fra_pop,icl_pop,nrw_pop)
print(type(kor_pop))
```

```python
kor_rate = round((usa_pop / kor_pop),2)
fra_rate = round((usa_pop / fra_pop),2)
icl_rate = round((usa_pop / icl_pop),2)
nrw_rate = round((usa_pop / nrw_pop),2)

print(kor_rate,fra_rate,icl_rate,nrw_rate)
kor.head()
```

```python
kor['KOR'] = kor['KOR'] * kor_rate
fra['FRA'] = fra['FRA'] * fra_rate
nrw['NRW'] = nrw['NRW'] * nrw_rate
icl['ICL'] = icl['ICL'] * icl_rate
```

```python
kor.head(20)
```

```python
from functools import reduce
```

```python
kor2 = kor[['date','KOR']]
usa2 = usa[['date','USA']]
fra2 = fra[['date','FRA']]
nrw2 = nrw[['date','NRW']]
icl2 = icl[['date','ICL']]

world = [kor2, usa2, fra2, nrw2, icl2]
```

```python
result = reduce(lambda left, right: pd.merge(left, right, on='date'), world).set_index('date')
result.head()
```

```python
df = pd.read_csv('/content/us-covid19.csv')
cols = ['date_updated', 'state', 'new_cases']

df = df[cols]
df.head()
```

```python
ca = df[df['state']=='CA']
ca = ca.rename(columns={'new_cases':'CA'})
ca.head()

il = df[df['state']=='IL']
il = il.rename(columns={'new_cases':'IL'})
il.head()

tx = df[df['state']=='TX']
tx = tx.rename(columns={'new_cases':'TX'})
tx.head()

wa = df[df['state']=='WA']
wa = wa.rename(columns={'new_cases':'WA'})
wa.head()

va = df[df['state']=='VA']
va = va.rename(columns={'new_cases':'VA'})
va.head()
```

```python
ca_pop = 39240000
il_pop = 12670000
tx_pop = 29530000
wa_pop = 7739000
va_pop = 8642000
usa_pop = 331900000


ca_rate = round((usa_pop / ca_pop),2)
il_rate = round((usa_pop / il_pop),2)
tx_rate = round((usa_pop / tx_pop),2)
wa_rate = round((usa_pop / wa_pop),2)
va_rate = round((usa_pop / va_pop),2)

print(ca_rate, il_rate, tx_rate, wa_rate, va_rate)
```

```python
ca['CA'] = ca['CA'] * ca_rate
ca.head()
```

```python
wa['WA'] = wa['WA'] * wa_rate


il['IL'] = il['IL'] * il_rate


tx['TX'] = tx['TX'] * tx_rate


va['VA'] = va['VA'] * va_rate
va.head(10)
```

```python
ca['date_updated'] = pd.to_datetime(ca['date_updated'])
ca = ca.set_index('date_updated')


il['date_updated'] = pd.to_datetime(il['date_updated'])
il = il.set_index('date_updated')


tx['date_updated'] = pd.to_datetime(tx['date_updated'])
tx = tx.set_index('date_updated')


wa['date_updated'] = pd.to_datetime(wa['date_updated'])
wa = wa.set_index('date_updated')


va['date_updated'] = pd.to_datetime(va['date_updated'])
va = va.set_index('date_updated')
```

```python
USAS = (ca,il,tx,wa,va)
result = reduce(lambda left, right : pd.merge(left, right, on='date_updated'), USAS).set_index('date_updated')
result.head(20)
```
